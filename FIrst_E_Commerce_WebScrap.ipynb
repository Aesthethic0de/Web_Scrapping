```python
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
import requests
```

```python
url = 'https://www.flipkart.com/layasa-sports-shoes-men-latest-stylish-casual-sneakers-lace-up-lightweight-running-walking-gym-trekking-hiking-party-running/product-reviews/itmaceeb358dd715?pid=SHOFWH3TXWRJS7EG&lid=LSTSHOFWH3TXWRJS7EGPQA0DT&marketplace=FLIPKART'
```

```python
page =  requests.get(url)
```

```python
print(page)
```

```python
page.content
```

```python
soup = bs(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('p', class_ = '_2sc7ZR _2V5EHH _1QgsS5')
```

```python
names
```

```python
cust_name = []
for i in range(0, len(names)):
  cust_name.append(names[i].get_text())
```

```python
cust_name
```

```python
revi = soup.find_all('div', class_ = '_6K-7Co')
```

```python
revi
```

```python
revi_1 = []
for i in range(0, len(revi)):
  revi_1.append(revi[i].get_text())
```

```python
revi_1
```

```python
dataf = pd.DataFrame()
dataf['Customer Name'] =  cust_name
dataf['reviews'] =  revi_1
```

```python
dataf
```

Lets Scraps Another E-Commerce Website for TITLE, PRICE AND REVIEW



```python
url_1 = 'https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
page1 =  requests.get(url_1)
```

```python
page1
```

```python
page1.content
```

```python
soup1 = bs(page1.content, 'html.parser')
```

```python
soup1
```

```python
print(soup1.prettify())
```

```python
iphone_names = soup1.find_all('div', class_ = '_4rR01T')
```

```python
iphone_names
```

```python
iphone_names_1 = []
for i in range(0, len(iphone_names)):
  iphone_names_1.append(iphone_names[i].get_text())
```

```python
iphone_names_1
```

```python
iphone_price = soup1.find_all('div', class_ = '_30jeq3 _1_WHN1')
```

```python
iphone_price
```

```python
iphone_price_1 = []
for i in range(0, len(iphone_price)):
 iphone_price_1.append(iphone_price[i].get_text())
```

```python
iphone_price_1
```

```python
iphone_reviews = soup1.find_all('span', class_ = '_2_R_DZ')
```

```python
iphone_reviews
```

```python
iphone_reviews_1 = []
for i in range(0, len(iphone_reviews)):
 iphone_reviews_1.append(iphone_reviews[i].get_text())
```

```python
iphone_reviews_1
```

```python
iphone_reviews_1[0].split()
```

```python
list_r_1 = []

for i in iphone_reviews_1:
  
  list_r_1.append(i.split())
```

```python
new_v = []
for i in list_r_1:
  rv = i[-1:-3:-1][::-1]
  print(rv)

  new_v.append(rv[0])
```

```python
new_v
```

```python
# new_v = []
# for i in list_r_1:
#   print(i[3:])
#   new_v.append(i[3:])
```

```python
dataframe_1 = pd.DataFrame()
dataframe_1['Product Name'] = iphone_names_1
dataframe_1['PRICE'] = iphone_price_1
dataframe_1["Reviews"] = new_v
```

```python
dataframe_1
```
